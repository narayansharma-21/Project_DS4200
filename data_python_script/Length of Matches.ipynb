```python
import altair as alt
import pandas as pd

length = pd.read_excel('All Time FV Match Length-2023-01-25-10-46-13.xlsx', header=8)
length = length.drop(length.columns[[0,2]], axis=1)
length = length.iloc[:-6]
length = length[(length['Length of Match (Months)'] > 0)]  
length = length.dropna()


length
```

```python
counts = length['Length of Match (Months)'].value_counts().reset_index(name='Num of Matches')
counts = counts.rename({'index': 'Match Length (Months)'}, axis=1)
counts
```

```python
match = alt.Chart(counts, title='Count of Length of Matches').mark_bar(size=10).encode(
    x = alt.X('Match Length (Months)', scale=alt.Scale(domain=[20, 200])),
    y = alt.Y('Num of Matches', scale=alt.Scale(domain=[0, 200])),
    color = alt.Color('Num of Matches'),
    tooltip =['Match Length (Months)' ,'Num of Matches']
).properties(
    width=2500
)

match.save("match.html")
match
```
